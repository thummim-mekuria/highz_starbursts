```python
ms = []
for i in range(201):
    try:
        particle_files[i].particles_exist
    except:
        ms.append(0*yt.units.solMass)
    else:
        ms.append(particle_files[i].all_data()['particle_mass'])
    finally:
        print(ms[i])
```

```python
particle_mass_tot = [np.nansum(ms[i]).in_units('solMass') for i in range(201)]
```

```python
plot_files = [yt.load("RHDTurb_hdf5_plt_cnt_{:04d}".format(n)) for n in range(201)]
```

```python
gas_mass = [np.nansum(plot_files[i].all_data()['mass']).in_units('solMass') for i in range(201)]
```

```python
ts = [file.current_time.in_units('yr') for file in plot_files]
```

```python
plt.rcParams['font.size'] = 15
plt.rcParams['font.family'] = 'STIXGeneral'
plt.rcParams['mathtext.fontset'] = 'stix'
```

```python
plt.plot(ts,gas_mass,marker='.',label='gas',linewidth=0.1)
plt.plot(ts,particle_mass_tot,marker='*',label='stars',linewidth=0.1)
plt.loglog()
plt.legend()
plt.xlabel('time [yr]')
plt.ylabel(r'mass [$M_\odot$]')
```

```python
init_gas = gas_mass[0]
```

```python
# plt.plot(ts,gas_mass,marker='.',label='gas',linewidth=0.1)
plt.plot(ts,particle_mass_tot/init_gas,marker='*',label='% of gas in stars',linewidth=0.1)
plt.loglog()
plt.legend()
plt.xlabel('time [yr]')
plt.ylabel(r'mass [$M_\odot$]')
```
